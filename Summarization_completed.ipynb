## Summarization
- concept of capturing very important gist of a long piece of text

### Types of Summarization
- 1. **Extractive Summarization**
    - Select sentences from the corpus that best represent the text
    - Arrange them to form a summary
- 2. **Abstractive Summarization**
    - Captures the very important sentences from the text
    - Paraphrases them to form a summary

### Import the library

```python
# Note: gensim 4.x removed gensim.summarization.
# We use gensim's TF-IDF model directly for extractive summarization.
import re
from gensim import corpora, models
```

### Scrape the text
- Use beautifulSoup to extract text (from Task1 of ML-1)

```python
import requests
from bs4 import BeautifulSoup

def get_page(url):
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(response.text, 'lxml')
    return soup
```

```python
def collect_text(soup):
    paragraphs = soup.find_all('p')
    text = ' '.join(p.get_text() for p in paragraphs if len(p.get_text().strip()) > 30)
    return text
```

```python
url = "https://en.wikipedia.org/wiki/Automatic_summarization"
```

```python
# Fetch and extract text from the page
soup = get_page(url)
text = collect_text(soup)
```

### Summarize
- **word_count**: maximum amount of words we want in the summary
- **ratio**: fraction of sentences in the original text should be returned as output

```python
def gensim_summarize(text, num_sentences=5):
    """Extractive summarization using Gensim TF-IDF scoring."""
    # Split text into sentences
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    sentences = [s for s in sentences if len(s.split()) > 5]

    # Tokenize each sentence (remove stopwords and short tokens)
    stop_words = {'the','a','an','is','are','was','were','it','its','of','in','to','and',
                  'or','for','on','at','by','as','be','with','this','that','from','not','also'}
    tokenized = [
        [w.lower() for w in s.split() if w.isalpha() and w.lower() not in stop_words and len(w) > 2]
        for s in sentences
    ]

    # Build dictionary and corpus
    dictionary = corpora.Dictionary(tokenized)
    corpus     = [dictionary.doc2bow(tokens) for tokens in tokenized]

    # Apply TF-IDF model and score sentences
    tfidf  = models.TfidfModel(corpus)
    scores = [(sum(weight for _, weight in tfidf[corpus[i]]), i)
              for i in range(len(sentences))]

    # Pick top-N sentences, preserving original order
    scores.sort(reverse=True)
    top_indices = sorted([idx for _, idx in scores[:num_sentences]])

    return [sentences[i] for i in top_indices]

gensim_summary = gensim_summarize(text, num_sentences=5)
print("Gensim TF-IDF Summary:")
for sentence in gensim_summary:
    print("-", sentence)
```

### Method 5: Gensim — TF-IDF Extractive Summarizer

```python
import re
from gensim import corpora, models

def gensim_summarize(text, num_sentences=5):
    """Extractive summarization using Gensim TF-IDF scoring."""
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    sentences = [s for s in sentences if len(s.split()) > 5]

    stop_words = {'the','a','an','is','are','was','were','it','its','of','in','to','and',
                  'or','for','on','at','by','as','be','with','this','that','from','not','also'}
    tokenized = [
        [w.lower() for w in s.split() if w.isalpha() and w.lower() not in stop_words and len(w) > 2]
        for s in sentences
    ]

    dictionary = corpora.Dictionary(tokenized)
    corpus     = [dictionary.doc2bow(tokens) for tokens in tokenized]
    tfidf      = models.TfidfModel(corpus)

    scores = [(sum(weight for _, weight in tfidf[corpus[i]]), i)
              for i in range(len(sentences))]
    scores.sort(reverse=True)
    top_indices = sorted([idx for _, idx in scores[:num_sentences]])

    return [sentences[i] for i in top_indices]

gensim_result = gensim_summarize(ARTICLE_TEXT, num_sentences=5)
print("=== Gensim TF-IDF Summary ===")
for i, sentence in enumerate(gensim_result, 1):
    print(f"{i}. {sentence}")
```
